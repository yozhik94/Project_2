Расчитали соотношение коэффициента дефолта в группе и целевой переменной (global_mean).
В статистики соотношение между вероятностями в разных группах называется коэффициентом риска, где риск относится к риску возникновения эффекта. В нашем случае эффект - дефолт клиента банка по кредиту.

Если риск ниже 1, то группа имеет более низкий уровень риска: уровень вероятности дефолта в этой группе меньше, чем глобальный дефолт (в нашем сучае global_mean = 3,5%, risk = 1).

- в 6 раз выше в enc_loans_credit_status (статус кредита) в значениях 6


```python
def plot_categorical_distributions(data, categorical_features, flag_column='flag', cols=2):
    from plotly.subplots import make_subplots
    import plotly.graph_objects as go
    import numpy as np

    # Определяем количество строк и столбцов для субплотов
    num_rows = (len(categorical_features) + cols - 1) // cols
    fig = make_subplots(
        rows=num_rows, cols=cols,
        subplot_titles=[f'Distribution of {feat} by Flag' for feat in categorical_features]
    )

    # Обходим все признаки и создаем графики
    for i, feature in enumerate(categorical_features):
        row = i // cols + 1
        col = i % cols + 1

        # Фильтруем данные по флагу
        flag_0_data = data[data[flag_column] == 0][feature]
        flag_1_data = data[data[flag_column] == 1][feature]

        # Получаем уникальные значения для признака
        unique_values = sorted(data[feature].unique())
        width = 0.2  # Ширина столбцов

        # Подсчет частот для уникальных значений
        hist0 = [flag_0_data.tolist().count(val) for val in unique_values]
        hist1 = [flag_1_data.tolist().count(val) for val in unique_values]

        # Создаем bar-графики для каждой категории
        fig.add_trace(
            go.Bar(
                x=unique_values,
                y=hist0,
                width=[width] * len(unique_values),
                name=f'{feature} - Flag 0',
                marker_color='blue',
                opacity=0.6,
                text=[f'{(h / len(flag_0_data)):.1%}' if h > 0 else '' for h in hist0],
                textposition='outside'
            ),
            row=row, col=col
        )

        fig.add_trace(
            go.Bar(
                x=[val + width for val in unique_values],
                y=hist1,
                width=[width] * len(unique_values),
                name=f'{feature} - Flag 1',
                marker_color='red',
                opacity=0.6,
                text=[f'{(h / len(flag_1_data)):.1%}' if h > 0 else '' for h in hist1],
                textposition='outside'
            ),
            row=row, col=col
        )

    # Обновление общей компоновки графика
    fig.update_layout(
        height=num_rows * 600,
        width=2000,
        title_text="Категориальные распределения по признакам",
        showlegend=True
    )

    # Обновление осей
    for i, feature in enumerate(categorical_features):
        fig.update_xaxes(title_text=feature, row=i // cols + 1, col=i % cols + 1)
        fig.update_yaxes(title_text="Частота", row=i // cols + 1, col=i % cols + 1)

    fig.show()
```

```python
# Вызов функции для построения графиков
plot_categorical_distributions(df_int8, binary_columns, flag_column='flag', cols=2)
```

## Under-sampling - уменьшение количества объектов доминирующего класса для отбора признаков по значимости

```python
X = df_int8.drop(['flag', 'rn', 'id'], axis=1)

y = df_int8['flag']
```

```python
del df_int8
```

```python
y.value_counts()
```

Instance Hardness Threshold

- Строит предсказание на основе модели классификации с использованием кросс-валидации (по умолчанию 5 фолдов).
- Оставляет правильно предсказанные объекты major-класса с наиболее высокими вероятностями предсказания.
- Могут использоваться различные модели классификации, по умолчанию применяется случайный лес.

```python
iht_hist = InstanceHardnessThreshold(
    estimator=HistGradientBoostingClassifier(random_state=42),
    sampling_strategy='auto',
    random_state=42,
    n_jobs=-1, cv=3)
hist_X, hist_y = iht_hist.fit_resample(X, y)
```

```python
hist_y.value_counts()
```

```python
hist_y.to_csv("/content/drive/MyDrive/Ml_Junior/Финальная работа/Итоговая работа_все файлы/ML_уменьшение_отбор/flag_hist.csv", index=False)
```

```python
X_train, X_test, y_train, y_test = train_test_split(hist_X, hist_y, test_size=0.2, random_state=42)
```

```python
model_cat = CatBoostClassifier(random_state=42, eval_metric='F1',
                               iterations=200, custom_metric=['F1', 'AUC'], depth=4,
                                early_stopping_rounds=20)
model_cat.fit(X_train, y_train, verbose=0)

# 5. Предсказание на отложенных данных
y_pred_train = model_cat.predict(X_train)
y_pred_cat = model_cat.predict(X_test)

# 6. Расчёт метрик качества
print("Классификационный отчет train:\n", classification_report(y_train, y_pred_train))
print("Confusion Matrix train:\n", confusion_matrix(y_train, y_pred_train))
print("Классификационный отчет test:\n", classification_report(y_test, y_pred_cat))
print("Confusion Matrix test:\n", confusion_matrix(y_test, y_pred_cat))

# ROC AUC
roc_auc_train = roc_auc_score(y_train, model_cat.predict_proba(X_train)[:, 1])
roc_auc_test= roc_auc_score(y_test, model_cat.predict_proba(X_test)[:, 1])
print("ROC AUC train:", roc_auc_train)
print("ROC AUC test:", roc_auc_test)

print('Точность для обучающей выборки:', model_cat.score(X_train, y_train))
print('Точность для тестовой выборки:', model_cat.score(X_test, y_test))
```

## Используем библиотеку SHAP для отбора признаков.

Отбор признаков проводила на общем файле без разделения на обучаемую и отложенную выборки.

Значения Шепли применяются в экономике, а точнее — в теории кооперативных игр. Такие значения назначаются игрокам сообразно их вкладу в игру. В сфере машинного обучения идея использования значений Шепли нашла отражение во фреймворке SHAP (SHapley Additive exPlanations).

Permutation Explainer не зависит от модели, поэтому он может вычислять значения Шепли для любой модели. Он работает, итерируя полные перестановки признаков вперед и назад.


```python
import shap
# построить Permutation Explainer и объяснить предсказания модели на заданном наборе данных
explainer = shap.explainers.Permutation(model_cat.predict_proba, X_test)
shap_values = explainer(X_test[:50000])

# получить только объяснения для положительного класса
shap_values = shap_values[..., 1]
```

```python
shap.plots.bar(shap_values, max_display=59)
```

Значение Оуэна.

В то время как значения Шепли получаются в результате обработки каждой функции независимо от других функций, часто бывает полезно навязать структуру входным данным модели. Навязывание такой структуры создает структурную игру (т. е. игру с правилами о допустимых коалициях входных функций), и когда эта структура является гнездовым набором группирования функций, мы получаем значения Оуэна как рекурсивное применение значений Шепли к группе. В SHAP строится бинарное иерархическое кластерное дерево для представления структуры данных. Эту структуру можно выбрать разными способами, но для табличных данных часто бывает полезно построить структуру на основании связи между входными функциями и целевой переменной(в нашем случау "flag").

```python
# построить кластеризацию признаков на основе общей информации о y_test
clustering = shap.utils.hclust(X_test, y_test)
```

```python
# теперь мы явно используем  Partition, который использует кластеризацию, которую мы только что вычислили
masker = shap.maskers.Partition(X_test, clustering=clustering)

# построить Permutation Explainer и объяснить предсказания модели на заданном наборе данных
explainer = shap.explainers.Permutation(model_cat.predict_proba, masker)
shap_values2 = explainer(X_test[:50000])


# получить только объяснения для положительного класса
shap_values2 = shap_values2[..., 1]
```

```python
shap.plots.bar(shap_values2, max_display=59)
```
